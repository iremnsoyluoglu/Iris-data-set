```python
#Görselleştirme / scatter
plt.figure()
plt.scatter(df.sepal_length,df.sepal_width,color = "orange",alpha = 0.9)
plt.title("iris sepal")
plt.xlabel("sepal lenth")
plt.ylabel("sepal width")
plt.grid(True)
plt.show()
```

```python
#Histogram
plt.figure() 
plt.hist(df["sepal_length"], color = "red", alpha = 0.5, bins = 20, label = "Alt Yaprak Uzunluk") 
plt.hist(df["petal_length"], color = "black", alpha = 0.5, bins = 20, label = "Üst Yaprak Uzunluk") 
plt.title('Alt Yaprak Uzunluğu ve Üst Yaprak Uzunluğu - Histogram')
plt.xlabel("Cm")
plt.ylabel("Frekans")
plt.legend()
plt.show()
```

```python
#Çizgi grafik
plt.figure() 
plt.plot(df["sepal_length"], color = "red", alpha = 0.5, label = "Alt Yaprak Uzunluk") 
plt.xlabel("Cm")
plt.ylabel("Frekans")
plt.legend()
plt.show()
```

```python
#Yaprak karşılaştırma
fig, ax = plt.subplots(2, 1) 
fig.suptitle("Alt Yaprak Uzunluk ve Üst Yaprak Uzunluk", fontsize=14)
ax[0].plot(df.sepal_length,color="green",label= "SepalLengthCm")
ax[0].set_title('Alt Yaprak Uzunluk') 
ax[0].set_xlabel("sepal") 
ax[0].set_ylabel("Cm") 

ax[1].plot(df.petal_length,color="yellow",label= "PetalLengthCm")
ax[1].set_title('Üst Yaprak Uzunluk')
ax[1].set_xlabel("petal")
ax[1].set_ylabel("Cm")

fig.tight_layout()
fig.subplots_adjust(top=0.85)
plt.show()
```

```python
# Pie grafik
species_counts = df["species"].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%',  colors=["orange", "pink", "yellow"], startangle=90, explode=(0.1, 0, 0))

plt.title("Çiçek Türlerinin Dağılımı")
plt.show()
```
